```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path1 = "/content/drive/MyDrive/python practising/athlete_events.csv"
```

```python
path2 = "/content/drive/MyDrive/python practising/noc_regions.csv"
```

```python
athletes = pd.read_csv(path1)
regions = pd.read_csv(path2)
print(regions.head(10))
```

```python
athletes.head()
```

```python
athletes_df = athletes.merge(regions, how = 'left', on = 'NOC')
athletes_df.head()
```

```python
athletes_df.rename(columns={'region':'Regions','notes':'Notes'})
```

```python
athletes_df.info()
```

```python
athletes_df.isnull().sum()
```

```python
athletes_df.query('Team=="India"').head(10)
```

```python
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
top_10_countries
```

```python
plt.figure(figsize=(12,6))
plt.title('Top 10 countries')
sns.barplot(x = top_10_countries.index, y=top_10_countries, palette = 'Set3')
```

```python
plt.figure(figsize=(12,6))
plt.title('Participants by Age')
plt.xlabel('Age')
plt.ylabel('Total participants')
plt.hist(athletes_df.Age, bins = np.arange(10,80,2), color = 'orange', edgecolor='white')
```

```python
winter_sports = athletes_df[athletes_df.Season =='Winter'].Sport.unique()
winter_sports
```

```python
gender_counts = athletes_df.Sex.value_counts()
gender_counts
```

```python
plt.figure(figsize=(12,6))
plt.title('Gender percentage')
plt.pie(gender_counts, labels = gender_counts.index, autopct='%1.1f%%')
```

```python
Women_athletes = athletes_df[athletes_df.Sex=='F'][['Medal','Year']]
Women_athletes = Women_athletes.groupby('Year').count().reset_index()
Women_athletes
```

```python
WomenOlympics = athletes_df[athletes_df.Sex=='F']
```

```python
sns.set(style="darkgrid")
plt.figure(figsize=(20,10))
plt.title('Women Participants By Year')
sns.countplot(x='Year', data = WomenOlympics, palette = "Spectral")
```
